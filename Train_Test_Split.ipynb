```python
#import relevant libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
#read dataset from excel
df = pd.read_excel('Colorectal Generated Data_New.xlsx')
df.set_index('Name',inplace=True)
```

```python
#separate cancer markers and input data
df_outputs= df['Marker']
df_inputs = df.drop('Marker',axis=1)
```

```python
#perform train-test split

train_inputs, test_inputs, train_outputs, test_outputs = train_test_split(df_inputs, df_outputs, test_size=0.2, shuffle=True)
```

```python
#export generated data split to excel 
with pd.ExcelWriter('Train_test_split.xlsx') as writer:
    #export train inputs and outputs to excel
    train_inputs.to_excel(writer, sheet_name='X_train', index = True)
    train_outputs.to_excel(writer, sheet_name='Y_train', index = True)

    #export test inputs and outputs to excel
    test_inputs.to_excel(writer, sheet_name='X_test', index = True)
    test_outputs.to_excel(writer, sheet_name='Y_test', index = True)
```

```python
#check that the file is being exported correctly
train_inputs = pd.read_excel('Train_test_split.xlsx', sheet_name='X_train')
train_inputs.set_index('Name',inplace=True)

train_outputs = pd.read_excel('Train_test_split.xlsx', sheet_name='Y_train')
train_outputs.set_index('Name',inplace=True)

test_inputs = pd.read_excel('Train_test_split.xlsx', sheet_name='X_test')
test_inputs.set_index('Name',inplace=True)

test_outputs = pd.read_excel('Train_test_split.xlsx', sheet_name='Y_test')
test_outputs.set_index('Name',inplace=True)

train_inputs.head()
```
